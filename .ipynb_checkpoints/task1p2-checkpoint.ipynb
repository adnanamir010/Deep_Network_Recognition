```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import cv2
import numpy as np
from networks import MyNetwork  # Make sure to import MyNetwork from networks.py
```

```python
def process_and_predict(image_path, model, device,invert):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Read the image in grayscale
    if invert:
        image = cv2.bitwise_not(image)
    image = cv2.resize(image, (28, 28))  # Resize to 28x28
    image = image.astype(np.float32) / 255.0  # Normalize image pixels to [0, 1]
    tensor = torch.tensor(image).unsqueeze(0).unsqueeze(0).to(device)  # Add batch and channel dimensions
    tensor = (tensor - 0.1307) / 0.3081  # Apply MNIST normalization
    output = model(tensor)
    return output
```

```python
def main():
    # Setup transformation and load the MNIST test dataset
    test_transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,))
    ])
    test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=test_transform)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=10, shuffle=False)

    # Initialize the network and load the pretrained weights from a .pt file
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = MyNetwork().to(device)
    model.load_state_dict(torch.load('mnist_model.pt', map_location=device))
    model.eval()

    # Get the first batch of 10 examples from the test set
    images, labels = next(iter(test_loader))
    outputs = model(images)  # Get model predictions

    # Print out the network output values, the index of the max output value, and the true label
    for idx, (output, label) in enumerate(zip(outputs, labels)):
        output_probs = torch.exp(output)  # Convert log softmax to probabilities
        print(f"Image {idx} - Predicted Label: {output_probs.argmax().item()}, True Label: {label.item()}")

        # Optionally, if you want to display the output values
        print(f"Output Values: {output_probs.detach().numpy()}")

    # Plot the first 9 images and their predictions
    plt.figure(figsize=(9, 9))
    for idx in range(9):
        plt.subplot(3, 3, idx+1)
        plt.imshow(images[idx][0], cmap='gray')
        plt.title(f"Predicted: {outputs[idx].argmax().item()}, True: {labels[idx]}")
        plt.axis('off')
    plt.show()

    print("My Custom Handwritten Digits")
    # Process and predict handwritten digits
    invert=False
    invert=!invert
    handwritten_digits_paths = ['0.png', '1.png', '2.png', '3.png', '4.png', '5.png','6.png','7.png','8.png','9.png']
    plt.figure(figsize=(10, 4))
    for i, image_path in enumerate(handwritten_digits_paths):
        output = process_and_predict(image_path, model, device,invert)
        prob = torch.exp(output)
        prediction = prob.argmax().item()
        plt.subplot(2, 5, i+1)
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if invert:
            image = cv2.bitwise_not(image)  # Invert the image to match MNIST
        plt.imshow(image, cmap='gray')
        plt.title(f'Predicted: {prediction}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
